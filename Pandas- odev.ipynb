5-Hangi ülkedekiler daha fazla bira içiyor ortalamada gösteriniz.

```python
# 5
a = df.groupby("country")["beer_servings"].mean()
pd.DataFrame(a.sort_values(ascending = False).head(10))  # en çok bira içen 10 ülke
```

6-Her kıta için şarap tüketiminin betimsel istatiğini yapınız.

```python
# 6
df.groupby("continent")["wine_servings"].describe()
```

7-Her kıta için ve her alkol cinsi için ortalama alkol tüketimini gösteriniz. 

```python
# 7
df.groupby(["continent"])[["wine_servings", "spirit_servings", "beer_servings"]].mean()
```

8-Her kıta için ve aşağıdaki alkol cinsleri için belirtilen hesaplamaları gerçekleştiriniz.

beer_servings - std
spirit_servings - median 
wine_servings - min, mean, max
total_litres_of_pure_alcohol - mean, count

```python
# 8
kosul = {"beer_servings": ["std"],
         "spirit_servings": ["median"],
         "wine_servings": ["min","mean","max"],
         "total_litres_of_pure_alcohol": ["mean","count"]}
df.groupby("continent").aggregate(kosul)
```

9- Her alkol cinsinin en çok tüketildiği ülkeyi bulun. 

```python
# 9
kosul = {"beer_servings": ["max"],
         "spirit_servings": ["max"],
         "wine_servings": ["max"]}
df.groupby("country").aggregate(kosul).sum(axis=1).sort_values(ascending=False).head(5) 
# En çok tüketen ülke Andorra
```

10- Yeni değişkenler üreterek her alkol cincisin ilgili ülkedeki toplam alkol tüketimi içerisindeki oranlarını yazdırınız.  

```python
# 10
for i in range(1, df.shape[0]+1):
    df["per_beer_servings"][[i-1]] =  (df["beer_servings"][i-1] / df[i-1:i][["beer_servings", "spirit_servings","wine_servings"]].sum(axis=1)).values[0]
    df["per_spirit_servings"][[i-1]] =  (df["spirit_servings"][i-1] / df[i-1:i][["beer_servings", "spirit_servings","wine_servings"]].sum(axis=1)).values[0]
    df["per_wine_servings"][[i-1]] =  (df["wine_servings"][i-1] / df[i-1:i][["beer_servings", "spirit_servings","wine_servings"]].sum(axis=1)).values[0]
df.fillna(0, inplace = True)
df
```
